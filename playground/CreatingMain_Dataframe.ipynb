# Importing IRS Data

```python
irs = pd.read_csv('./datasets/16zpallagi.csv')
irs.head() 
```

## Summarizing IRS data by zip code

```python
pivot_irs = irs.pivot_table(index = 'zipcode')
```

```python
pivot_irs.head()
```

```python
pivot_irs.shape
```

```python
pivot_irs.reset_index(level=0, inplace=True)
```

```python
pivot_irs.head()
```

```python
pivot_irs.shape
```

```python
#lets change zip code to string 
pivot_irs['zipcode'] = pivot_irs['zipcode'].astype(str)
```

```python
pivot_irs.dtypes.head()
```

## Keeping Zip Codes that only start with 9

```python
irs = pivot_irs[pivot_irs['zipcode'].astype(str).str.startswith('9')]
```

```python
irs.head()
```

```python
irs.shape
```

```python
irs.rename(columns = {'zipcode': 'zip_code'}, inplace= True)
```
