# Find Categories that have high correlation to Income

```python
categories_with_income = df.loc[:, 'acaibowls':'income']
```

```python
categories_with_income.shape
```

```python
categories_with_income.head()
```

```python
#creating list of category columns
list_categories = list(categories_with_income.columns)
```

```python
len(list_categories)
```

```python
#removing target variable from list 
list_categories.remove('income')
```

```python
len(list_categories)
```

```python
#creating for loop that calculates the correlation and keep the category names that have a high correlation to income
#positive correlation first 
pos_corr_col = []
for col in list(list_categories):
    corr = categories_with_income['income'].corr(categories_with_income[col])
    if corr > .15:
        pos_corr_col.append(col)
```

```python
pos_corr_col
```

```python
len(pos_corr_col)
```

```python
#negative correlation categories 
#creating for loop that calculates the correlation and keep the category names that have a high correlation to income

neg_corr_col = []
for col in list(list_categories):
    corr = categories_with_income['income'].corr(categories_with_income[col])
    if corr < -0.09:
        neg_corr_col.append(col)
```

```python
neg_corr_col
#we'll consider `salvadoran`
```

```python
correlation_cat = pos_corr_col+neg_corr_col
```

```python
correlation_cat.remove('restaurants')
```

```python
len(correlation_cat)
```

```python
correlation_cat
```

# Feature Engineering Number of Businesses based on Price and Rating 
- We want the percentage of number of businesses based on specific price and ratings

```python
df.head()
```

# Feature Engineering Price

```python
df['price_1_percentage'] = df['Price_1']/ df['count']
```

```python
#sanity check if it worked 
df.head()
```

```python
df['price_2_percentage'] = df['Price_2']/ df['count']
df['price_3_percentage'] = df['Price_3']/ df['count']
df['price_4_percentage'] = df['Price_4']/ df['count']
```

```python
df.head()
```

# Feature Engineering Ratings

```python
ratings_col = [col for col in df if col.startswith('Rating')]
ratings_col
```

```python
df['rate_1_percentage'] = df['Rating_1']/ df['count']
df['rate_1.5_percentage'] = df['Rating_1.5']/ df['count']
df['rate_2_percentage'] = df['Rating_2']/ df['count']

df['rate_2.5_percentage'] = df['Rating_2.5']/ df['count']
df['rate_3_percentage'] = df['Rating_3']/ df['count']
df['rate_3.5_percentage'] = df['Rating_3.5']/ df['count']

df['rate_4_percentage'] = df['Rating_4']/ df['count']
df['rate_4.5_percentage'] = df['Rating_4.5']/ df['count']
df['rate_5_percentage'] = df['Rating_5']/ df['count']
```
