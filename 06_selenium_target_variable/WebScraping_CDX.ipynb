```python
df = pd.read_csv('../datasets/MOTHEROFMOTHERS.csv')
df.head()
```

```python
#preparing list of zip codes for scraping median income based on zip code
zip_list = list(df['zip_code'])
```

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from time import sleep

url = 'https://www.cdxtech.com/tools/bulk/demographics/radius/'
```

```python
from selenium import webdriver

driver = webdriver.Chrome(executable_path="./chromedriver/macos/chromedriver") 
```

```python
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
```

```python
zip_list.index(98597)
```

```python
target = {}
for i in range(2175, len(zip_list)):
    
    #find the input box and enter specified zip 
    #find input box and enter 1 as radius in miles
    try: 
        inputElementZip = driver.find_element_by_id("Zipcode")
        inputElementZip.send_keys(str(zip_list[i]))
        inputElementMiles = driver.find_element_by_id("Mileage")
        inputElementMiles.send_keys(str(1)) #1st form of enter
        
        #when both values are inputed, click the button
        driver.execute_script("document.querySelectorAll('.btn.btn-primary')[1].click()")
        
        #wait 1.5 seconds once the page loads
        time.sleep(1.5)
        
        #once the page loads, locate median income 
        try:
            word =driver.find_element_by_xpath('//*[@id="PreviewTable"]/tbody/tr[1]/td[16]')
        #if it cant be located, use this path (since the web developers didnt add [1] for some of the zip codes)
        except NoSuchElementException:
            word =driver.find_element_by_xpath('//*[@id="PreviewTable"]/tbody/tr/td[16]')
    
    #if zip code cant be found in general, go onto the next zip code
    except NoSuchElementException:
        pass
    
    #if median income is scraped, append it to the target dictionary
    try:
        target[str(zip_list[i])] = word.text
    
    #if it did not collect median income, move onto the next zip code
    except StaleElementReferenceException:
        pass
    
    #go onto the main page to restart the search in order to find the median income for the next zip code and wait 1.5
    #seconds before proceeding to scrape 
    driver.execute_script('document.querySelectorAll(".btn.btn-primary.btn-lg")[0].click()')
    time.sleep(1.5)
```
